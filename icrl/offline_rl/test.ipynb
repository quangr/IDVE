```python
import matplotlib.pyplot as plt
from itertools import product
import numpy as np
p_lowers=[0.001,0.005,0.01,0.05]
lambs = [0.1,0.3,0.5,0.7,0.9]
max_constraints=[100,400,700,1000]
products = list(product(p_lowers, lambs))

fig = plt.figure(figsize=(40,5))

outter_fig = fig.subfigures(1, len(max_constraints))
for max_constraint,subfig in zip(max_constraints,outter_fig.flat):
    axs = subfig.subplots(len(p_lowers), len(lambs))
    for p_lower, lamb in products:
        key_string = f"p_lower_{p_lower}_lamb_{lamb}_max_constraint_{max_constraint}"
        cost=[]
        for data in datas:
            for key, value in data.items():
                if key_string in key:
                    cost.append(value[0])
        cost=np.array(cost)

        # Plot the heatmap
        cost=cost.mean(0)
        im = axs[p_lowers.index(p_lower),lambs.index(lamb)].imshow(cost, cmap="gray", alpha=0.9)

    for ax, col in zip(axs[0], lambs):
        ax.set_title(col)

    for ax, row in zip(axs[:,0], p_lowers):
        ax.set_ylabel(row, rotation=0, size='large')

    # fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from itertools import product
import numpy as np
p_lowers=[0.001,0.005,0.01,0.05]
lambs = [0.1,0.3,0.5,0.7,0.9]
max_constraints=[100,400,700,1000]
products = list(product(lambs, max_constraints))

fig = plt.figure(figsize=(8,4))
p_lower=p_lowers[1]
axs = fig.subplots(len(max_constraints),len(lambs) )
for lamb, max_constraint in products:
    key_string = f"p_lower_{p_lower}_lamb_{lamb}_max_constraint_{max_constraint}"
    cost=[]
    for data in datas:
        for key, value in data.items():
            if key_string in key:
                cost.append(value[0])
    cost=np.array(cost)

    # Plot the heatmap
    cost=cost.mean(0)
    im = axs[max_constraints.index(max_constraint),lambs.index(lamb)].imshow(cost, cmap="gray", alpha=0.9)
    axs[max_constraints.index(max_constraint),lambs.index(lamb)].tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labeltop=False, labelleft=False, labelright=False)
    # axs[max_constraints.index(max_constraint),lambs.index(lamb)].grid(True)
for ax, col in zip(axs[:,0], max_constraints):
    ax.set_ylabel(f"reg_eff={col}", rotation=0, size='large')
    ax.yaxis.set_label_coords(-1,0.5)
for ax, row in zip(axs[0], lambs):
    ax.set_title(f"lamb={row}")

    # fig.tight_layout()
plt.show()
```
